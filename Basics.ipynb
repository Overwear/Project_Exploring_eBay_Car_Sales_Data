## Next Steps

Now I want to take this a step further and use aggregation to explore the variations across different car brands. 

```python
#Checking for top 10 brands sold on the german eBay car listing
new_autos["brand"].value_counts().head(10)
```

### Next steps

I now want to see the oldest registration year by brand. Here I will use aggregation to manipulate data based on each existing unique brand.

```python
oldest_car_by_brand = {}
brands = autos["brand"].unique()

for brand in brands:
    selected_rows = new_autos[new_autos["brand"] == brand].sort_values("registration_year", ascending = True)
    oldest_car_row = selected_rows.iloc[0]
    oldest_car_by_brand[brand] = oldest_car_row["registration_year"]

oldest_car_by_brand
```

```python
#Get the oldest year of the car listed
min_year = min(oldest_car_by_brand.values())

#Get the brand of the oldest car listed
min_brand = min(oldest_car_by_brand, key = oldest_car_by_brand.get)

print("The minimum year is {0} and the associated brand is {1}".format(min_year,min_brand))
```

### Observations

One of the oldest car on the german eBay listing site was a 1910 trabant. 

I can expand further on this since I noticed that there are more vehicles with the registration year of 1910. The min function I used only returned the lowest year and the associated brand with that value that it last found. I could next create a list of brands that meet that date, but will do that in the future.

## Next Steps

Next I want to see the mean price of the top 10 brands sold on the website.

```python
#Get top ten brands sold 
top_ten_brands = new_autos["brand"].value_counts().head(10)

#Convert the brands to a list
top_ten_brands_list = top_ten_brands.index.tolist()
```

```python
#Create dictionary to hold the average price of the top 10 brands
top_10_average_price = {}

for brand in top_ten_brands_list:
    df_containing_brand_only = new_autos[new_autos["brand"] == brand]
    price_sum = df_containing_brand_only["price"].sum()
    how_many_cars_in_brand = df_containing_brand_only.shape[0]
    average_price = price_sum/how_many_cars_in_brand
    top_10_average_price[brand] = average_price
    print("The brand is {0} \nThe price_sum is {1} \nThe number of cars of this brand in this dataset is {2} \nThe average price for this brand is {3}\n ------------------------".format(brand, price_sum, how_many_cars_in_brand, average_price))
top_10_average_price
```

```python
highest_average_price = max(top_10_average_price.values())
highest_average_price_brand = max(top_10_average_price, key = top_10_average_price.get)
print("The highest average price is {0} and the associated brand is {1}".format(highest_average_price, highest_average_price_brand))
```

### Observations

The highest average price is 30317.48 and the associated brand is mercedes benz. This is actually quite high relative to the other top 10 averages. Future investigations would include finding out if there are any price outliers that are causing this to be so high.

I also found that the luxury brands such as audi, bmw and mercedes benz had a higher average price compared to the rest of the top 10.

## Next Steps

Now I want to get the mean milage of the cars and the prices and display them both in one data frame.


```python
#Create dictionary to hold the average milage of the top 10 brands
top_10_average_milage = {}

for brand in top_ten_brands_list:
    df_containing_brand_only = new_autos[new_autos["brand"] == brand]
    milage_sum = df_containing_brand_only["odometer_km"].sum()
    how_many_cars_in_brand = df_containing_brand_only.shape[0]
    average_milage = milage_sum/how_many_cars_in_brand
    top_10_average_milage[brand] = average_milage
    print("The brand is {0} \nThe milage_sum is {1}km \nThe number of cars of this brand in this dataset is {2} \nThe average milage for this brand is {3}km\n ------------------------".format(brand, milage_sum, how_many_cars_in_brand, average_milage))
top_10_average_milage
```

```python
mean_milage_series = pd.Series(top_10_average_milage)
mean_price_series = pd.Series(top_10_average_price)

new_mean_df = pd.DataFrame(mean_milage_series, columns=['mean_milage'])

new_mean_df['mean_price'] = mean_price_series

new_mean_df['KM/Dollar'] = new_mean_df['mean_milage'] / new_mean_df['mean_price']

new_mean_df
```

### Observations

Surprisingly the benz brand has a really high mean milage with the highest mean price. I am curious if there are outliers that are really messing with my data set.

I also added a new column that showed the average milage over the average price. To interpret this column.. anything on the higher end is great bang for your buck such as the renault with a score of 53.51. Anything on the lower end is a really expensive buy for the amount of milage already on that car.. for example the mercedes benz with a score of 4.316.
